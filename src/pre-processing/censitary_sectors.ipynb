```python
import pandas as pd
import geopandas as gpd
pd.set_option('display.max_columns', None)
```

# SETORES 2010

```python
sectors2010 = gpd.read_file('/Volumes/ssd/Doutorado/DATASET/IBGE/Setores Censitários/setores_censitarios_2010/br/setores_censitarios.shp')
sectors2010.head()
```

```python
ibp2010 = pd.read_csv('/Volumes/ssd/Doutorado/DATASET/IBP - CIDACS/ibp_setor_censitario-1.csv')
ibp2010
```

```python
sectors2010['CD_GEOCODI'] = sectors2010['CD_GEOCODI'].astype('Int64')
sectors2010 = sectors2010.merge(ibp2010, left_on='CD_GEOCODI', right_on='Cod_setor', how='left')
sectors2010
```

```python
sectors2010 = sectors2010[['ID', 'CD_GEOCODI', 'geometry', 'V001_p13', 'BrazDep_measure']]
sectors2010.columns = ['id', 'cod_setor', 'geometry', 'pop', 'brazdep']
sectors2010
```

```python
sectors2010.to_file('/Volumes/ssd/Doutorado/DATASET/IBGE/Setores Censitários/setores_censitarios_2010/br/setores_censitarios_pop_ibp.shp')
```

# SETORES 2022

```python
sectors2022 = gpd.read_file('/Volumes/ssd/Doutorado/DATASET/IBGE/Setores Censitários/setores_censitarios_2022/br/setores_censitarios_br.shp')
sectors2022.head()
```

```python
agg_sectors = pd.read_csv('/Volumes/ssd/Doutorado/DATASET/IBGE/Setores Censitários/Agregados_por_setores_basico_BR_2022.csv', sep=';', encoding='latin-1')
agg_sectors
```

```python
sectors2022['CD_SETOR'] = sectors2022['CD_SETOR'].astype('Int64')
sectors2022 = sectors2022.merge(agg_sectors, left_on='CD_SETOR', right_on='CD_SETOR', how='left')
sectors2022
```

```python
sectors2022 = sectors2022[['CD_SETOR', 'geometry', 'v0001', 'CD_FCU_x']]
sectors2022.columns = ['cod_setor', 'geometry', 'pop', 'cd_fcu']
sectors2022
```

```python
sectors2022.to_file('/Volumes/ssd/Doutorado/DATASET/IBGE/Setores Censitários/setores_censitarios_2022/br/setores_censitarios_pop_fcu.shp')
```
