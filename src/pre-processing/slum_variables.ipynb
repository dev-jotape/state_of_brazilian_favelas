# WITHIN AREA LEVEL

## Physical Hazards & Asset

### Ecological diversity

```python
mapbiomas2023 = gpd.read_file('/Volumes/ssd/Doutorado/DATASET/IBGE/Favelas e Comunidades Urbanas - 2022/poligonos/Mapbiomas/mapbiomas_2023.shp')
mapbiomas2023 = mapbiomas2023[mapbiomas2023['area_favel'] > 0]
mapbiomas2023

# Vegetation area
# Water area
# porcentagem da area
mapbiomas2023['wtr_area'] = (mapbiomas2023['agua'] / mapbiomas2023['area_favel']) * 100
mapbiomas2023['veg_area'] = ((mapbiomas2023['veg_intra_'] + mapbiomas2023['frag_veg']) / mapbiomas2023['area_favel']) * 100

ecological = mapbiomas2023[['cd_fcu', 'wtr_area', 'veg_area']]
ecological.columns = ['CD_FCU', 'wtr_area', 'veg_area']

final_table = final_table.merge(ecological, on='CD_FCU', how='left')
```

### Natural hazards (slope, flood zone)

```python
# Risk area
# Declive area
# HAND
mapbiomas2023['risk'] = (mapbiomas2023['risco'] / mapbiomas2023['area_favel']) * 100
mapbiomas2023['slope_10'] = (mapbiomas2023['decliv_10'] / mapbiomas2023['area_favel']) * 100
mapbiomas2023['slope_20'] = (mapbiomas2023['decliv_20'] / mapbiomas2023['area_favel']) * 100
mapbiomas2023['slope_30'] = (mapbiomas2023['decliv_30'] / mapbiomas2023['area_favel']) * 100
mapbiomas2023['slope_30p'] = (mapbiomas2023['decliv_mai'] / mapbiomas2023['area_favel']) * 100
mapbiomas2023['hand_6p'] = (mapbiomas2023['hand_acima'] / mapbiomas2023['area_favel']) * 100
mapbiomas2023['hand_0_3'] = (mapbiomas2023['hand_ate_3'] / mapbiomas2023['area_favel']) * 100
mapbiomas2023['hand_3_6'] = (mapbiomas2023['hand_ate_6'] / mapbiomas2023['area_favel']) * 100


natural_hazards = mapbiomas2023[['cd_fcu', 'risk', 'slope_10', 'slope_20', 'slope_30', 'slope_30p', 'hand_6p', 'hand_3_6', 'hand_0_3']]
natural_hazards.columns = ['CD_FCU', 'risk_area', 'slope_10', 'slope_20', 'slope_30', 'slope_30p', 'hand_6p', 'hand_3_6', 'hand_0_3']

final_table = final_table.merge(natural_hazards, on='CD_FCU', how='left')
final_table
```
