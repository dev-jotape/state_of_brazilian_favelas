```python
import pandas as pd
import geopandas as gpd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gdf = gpd.read_file('../../dataset/shp/final_version_no_missing_values.shp')
gdf.columns
```

```python
def plot_data(vars_x, vars_y):
    corr_matrix = pd.DataFrame(index=vars_x, columns=vars_y)

    for x in vars_x:
        for y in vars_y:
            corr_matrix.loc[x, y] = gdf[x].corr(gdf[y])

    # Converter valores para float
    corr_matrix = corr_matrix.astype(float)

    # Plotar heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f")
    plt.title("Correlação entre variáveis demográficas e indicadores")
    plt.xlabel("Indicadores")
    plt.ylabel("Variáveis demográficas")
    plt.tight_layout()
    plt.show()
```

## S.E.S (socio-economic status)

```python
vars_x = ['population', 'density_km', 'age_idx', 'median_age', 'female', 'literate', 'race_wht', 'race_blk', 'race_ylw', 'race_brn', 'race_ind', 'race_unk', 'depriv_idx']
vars_y = ['wtr_idx', 'sanit_idx', 'liv_a_idx', 'hh_idx']
plot_data(vars_x, vars_y)
```
